```python
import pandas as pd
```

```python
df = pd.read_csv('./podcast_db.csv')
df.head()
```

```python
import concurrent.futures
import os
```

```python
from pypdf import PdfReader

def extract_text(file_path):
    reader = PdfReader(file_path)
    text = ''
    for page in reader.pages:
        text += page.extract_text()
    return text
```

```python
def convert_pdf_to_text(pdf_path, save_path):
    text = extract_text(pdf_path)
    with open(save_path, 'w') as f:
        f.write(text)
```

```python
TXT_DIR = './pdfs_txt/'

if not os.path.exists(TXT_DIR):
    os.makedirs(TXT_DIR)
```

```python
def process_pdf(row):
    pdf_path = row['file']
    save_path = TXT_DIR + pdf_path.split('/')[-1].replace('.pdf', '.txt')
    return pdf_path, save_path
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=os.cpu_count()) as executor:
    futures = []
    for i, row in df.iterrows():
        pdf_path, save_path = process_pdf(row)
        future = executor.submit(convert_pdf_to_text, pdf_path, save_path)
        futures.append(future)

    for future in concurrent.futures.as_completed(futures):
        try:
            future.result()
        except Exception as e:
            print("Error processing pdf:", e)
```

```python
print('Done')
```

```python
SAVE_DIR = './pdfs_txt/'

total_files_saved = 0

for i, row in df.iterrows():
    pdf_path, save_path = process_pdf(row)
    
    ## check if the file exists
    if os.path.exists(save_path):
        total_files_saved += 1
        
assert total_files_saved == len(df)
```
