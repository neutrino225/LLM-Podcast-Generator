```python
import requests

def download_arxiv_paper(arxiv_id, save_path):
    url = f"https://arxiv.org/pdf/{arxiv_id}.pdf"
    response = requests.get(url)
    if response.status_code == 200:
        with open(save_path, 'wb') as f:
            f.write(response.content)
    else:
        print(f"Failed to download paper {arxiv_id}. Status code: {response.status_code}")
```

```python
import os
import requests

def download_arxiv_pdfs(paper_ids, save_dir="arxiv_papers"):
    """
    Downloads PDFs from arXiv given a list of paper IDs.
    
    Parameters:
    - paper_ids (list): List of arXiv paper IDs.
    - save_dir (str): Directory to save the downloaded PDFs.
    """
    # Create the directory if it doesn't exist
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    for paper_id in paper_ids:
        try:
            # Construct the PDF URL
            url = f"https://arxiv.org/pdf/{paper_id}.pdf"
            response = requests.get(url, stream=True)

            # Check if the request was successful
            if response.status_code == 200:
                # Save the PDF to the specified directory
                file_path = os.path.join(save_dir, f"{paper_id}.pdf")
                with open(file_path, "wb") as f:
                    for chunk in response.iter_content(chunk_size=1024):
                        f.write(chunk)
                print(f"Downloaded: {paper_id}")
            else:
                print(f"Failed to download {paper_id}: HTTP {response.status_code}")
        except Exception as e:
            print(f"Error downloading {paper_id}: {e}")

# Example usage
paper_ids = ["0704.0001", "0704.0002", "0704.0003"]  # Replace with your list of IDs
download_arxiv_pdfs(paper_ids)
```

## Create podcasts

```python
from pypdf import PdfReader

def extract_text(file_path):
    reader = PdfReader(file_path)
    text = ''
    for page in reader.pages:
        text += page.extract_text()
    return text
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM
import re
import os

def strip_markdown(text):
    # Remove Markdown headings, bold, italics, etc.
    clean_text = re.sub(r"([*_#>`\-\+])", "", text)
    return clean_text.strip()

template = """
You are an expert podcast generator. Your name is Sarah and the guest you have is Pete. You have been asked to create a podcast episode on the given pdf. The episode should be informative and engaging.

Here is the content of the pdf:
{content}

Please create a podcast episode on this topic without any headings and clearly distinguish between host and guest parts use HOST and GUEST. The audience is interested in learning more about this topic.
"""

prompt = ChatPromptTemplate.from_template(template)

# Replace <server-ip> and <port> with actual values if not running locally
model = OllamaLLM(model="llama3.2:3b", base_url="http://172.24.144.1:11434")

chain = prompt | model
```

```python
from tqdm import tqdm
import os

DIR = "./Data/arxiv"
PODCAST_DIR = "./Data/podcasts"

db = {
    "file": [],
    "podcast": []
}

# Initialize tqdm for progress tracking
for file in tqdm(os.listdir(DIR), desc="Processing Files"):
    filepath = os.path.join(DIR, file)
    save_path = os.path.join(PODCAST_DIR, file.replace(".pdf", ".txt"))

    # Check if the podcast file already exists
    if os.path.exists(save_path):
        db["file"].append(filepath)
        db["podcast"].append(save_path)
        continue

    try:
        # Extract text from the file
        text = extract_text(filepath)
        
        # Generate podcast script using your model or chain
        response = chain.invoke({'content': text})

        # Save the response to a text file
        with open(save_path, "w") as f:
            f.write(response)
            
        # Update database
        db["file"].append(filepath)
        db["podcast"].append(save_path)

    except Exception as e:
        print(f"Failed to process {file}: {e}")
        continue
```

```python
import pandas as pd
df = pd.DataFrame(db)
df.to_csv("podcast_db.csv", index=False)
```
